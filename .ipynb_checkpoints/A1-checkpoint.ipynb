```python
from pyspark.sql import SparkSession
from operator import add
from pyspark.ml.feature import Tokenizer
import re

# (8 cores, 16gb per machine) x 5 = 40 cores

# New API
spark_session = SparkSession\
        .builder\
        .master("spark://fabian-a2-from-snap:7077") \
        .appName("simple_example")\
        .getOrCreate()

# Old API (RDD)
spark_context = spark_session.sparkContext

data_frame = spark_session.read\
    .option("header", "true")\
    .text('europarl-v7.sv-en.sv')\
    .cache()
    
sv_lines = spark_context.textFile('europarl-v7.sv-en.sv').cache()
en_lines = spark_context.textFile('europarl-v7.sv-en.en').cache()
```

```python
tokenized = Tokenizer(inputCol="value",outputCol="tokenized").transform(data_frame)
#tokenized.filter(tokenized.value.like("%<%")).take(10)




```

```python
import re
```

```python
#Get the 10 most common w
en_lines\
.map(lambda x: x.lower())\
.flatMap(lambda t: t.split(' '))\
.flatMap(lambda w: w.split('\n'))\
.map(lambda w: w.strip()).cache()\
.map(lambda w: (w,1))\
.reduceByKey(add)\
.takeOrdered(10, key=lambda x: -x[1])
```

```python
#pair translation  - zip en with index
RDD_en = en_lines\
.map(lambda x: x.lower())\
.map(lambda x: re.sub(r'\W+', ' ', x))\
.map(lambda x: re.sub(r'[0-9]+', '', x))\
.flatMap(lambda w: w.split('\n'))\
.map(lambda w: w.strip())\
.zipWithIndex()\
.map(lambda z: (z[1], z[0]))
```

```python
#Zip sv with index
RDD_sv = sv_lines\
.map(lambda x: x.lower())\
.map(lambda x: re.sub(r'\W+', ' ', x))\
.map(lambda x: re.sub(r'[0-9]+', '', x))\
.flatMap(lambda w: w.split('\n'))\
.map(lambda w: w.strip())\
.zipWithIndex()\
.map(lambda z: (z[1], z[0]))\
```

```python
RDD_joined = RDD_en.join(RDD_sv)
RDD_joined.cache()
```

```python
#find suitable sentences/lines
max_length = 4

RDD_unzipped = RDD_joined\
.map(lambda t: (t[1][0].split(' '), t[1][1].split(' ')))\
.filter(lambda x: x[0][0] != "")\
.filter(lambda x: len(x[0]) < max_length)\
.filter(lambda t: len(t[0]) == len(t[1]))\

RDD_unzipped.take(400)
```

```python
#Make dictionaries out of suitable sentences and return the 10 most commons combinations

RDD_unzipped\
.flatMap(lambda t: t[0])\
.zip(RDD_unzipped.flatMap(lambda t: t[1]))\
.map(lambda w: (w,1))\
.reduceByKey(add)\
.takeOrdered(20, key=lambda x: -x[1])
```
